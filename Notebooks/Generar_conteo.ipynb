```python
import pandas as pd

reviews = pd.read_csv('kindle_reviews.csv')
# Filtrar reseñas con calificación 4 o 5
reviews[reviews['overall']>=4]
# Verificar que la columna 'reviewText' exista
if 'reviewText' not in reviews.columns:
    raise KeyError("'reviewText' column not found in the dataset")
# Normalizar a minúsculas para búsqueda insensible a mayúsculas
reviews['reviewText_lower'] = reviews['reviewText'].astype(str).str.lower()
# Filtrar por la raíz 'anxiet' para capturar 'anxiety', 'anxious', 'anxieties', etc.
mask = reviews['reviewText_lower'].str.contains(r'anxiet', na=False)
anxiety_reviews = reviews[mask].copy()
# Agrupar por 'asin' y contar ocurrencias, luego ordenar descendente
counts = anxiety_reviews.groupby('asin').size().reset_index(name='count').sort_values('count', ascending=False)
# Guardar resultados en CSV
counts.to_csv('anxiety_counts_by_asin.csv', index=False)
# Mostrar las top 50 ASINs con más menciones
counts.head(50)
```
